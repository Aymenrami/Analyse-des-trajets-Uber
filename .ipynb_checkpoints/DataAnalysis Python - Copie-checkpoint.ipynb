```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('UBERBigData.csv')
```

```python
X=df.iloc[:,:100000]
```

```python
sns.histplot(df['Jour de la semaine'],kde = True)
plt.xticks(rotation = 0)
plt.show()
```

```python
sns.histplot(df['Heure'],kde = True)
```

```python
g=sns.FacetGrid(df, col='DayType')
g.map(sns.histplot, 'Heure' , kde = True)
```

```python
count= df['DayType'].value_counts()
plt.pie(count ,startangle= 90 , autopct='%.0f%%',colors =['#2D3250','#F6B17A'] , shadow=True)
plt.legend(labels=count.index, loc='upper right', bbox_to_anchor=(1.1,1))
plt.show()
```

```python
corre = df.groupby(['Nom du jour' , 'Heure']).apply(lambda x: len(x))
corre=corre.unstack()
sns.heatmap(corre)
plt.show()
```
